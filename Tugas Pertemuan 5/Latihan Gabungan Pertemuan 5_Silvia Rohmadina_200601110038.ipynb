### 1. Buatlah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Program kemudian menampilkan penjumlahan N bilangan genap positif pertama (bilangan genap ≥ 0). Contoh

### • Jika user memasukkan N = 3, maka outputnya: 0 + 2 + 4 = 6

### • Jika user memasukkan N = 5, maka outputnya: 0 + 2 + 4 + 6 + 8 = 20

```python
print('Program Menampilkan Penjumlahan N Bilangan Genap Positif Pertama')

def pengulangan():
    N = int(input('Masukkan Nilai N :'))
    count = 0
    sum = 0
    for i in range (0,N*2,2):
        if i % 2 == 0:
            count = count + 1
            sum = sum + i
            print('Bilangan Genap Positif Ke -',count,'adalah',i)
    print()       
    print('Jika user memasukkan N:',N,'maka outputnya:',sum)
    
    print ('')
    print ('ingin coba lagi ? (ya/tidak)')
    x=input()
    if x=='ya':
        return pengulangan()
    if x=='tidak':
        print('Terimakasih! ')
        
pengulangan()
```

### 2. Buatlah sebuah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Kemudian, program menampilkan penjumlahan N bilangan kuadrat pertama. Bilangan kuadrat adalah = 1, 4, 9, 16, 25, 36, ...., N2

### Contoh:
    
### • Jika user memasukkan N = 2, maka outputnya: 1 + 4 = 5
    
### • Jika user memasukkan N = 3, maka outputnya: 1 + 4 + 9 = 14

```python
print('Program Menampilkan Penjumlahan N Bilangan Kuadrat Pertama')

def pengulangan():
    N = int(input('Masukkan Nilai N :'))
    count = 0
    sum = 0
    for i in range(1,N+1):
        if i > 0:
            count = count + 1
            sum = sum + i*i
            print('Bilangan Kuadrat Ke -',count, ' :',i*i)
    print()       
    print('Jika user memasukkan N:',N,'maka outputnya:',sum)
    
    print ('')
    print ('ingin coba lagi ? (ya/tidak)')
    x=input()
    if x=='ya':
        return pengulangan()
    if x=='tidak':
        print('Terimakasih!')
        
pengulangan()
```
